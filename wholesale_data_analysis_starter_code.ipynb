## Part 2: Transform the Data
Do we know that this client spent the more money than client 66037? If not, how would we find out? Transform the data using the steps below to prepare it for analysis.

```python
# Create a column that calculates the 
# subtotal for each line using the unit_price
# and the qty
wholesale_df['subtotal_line'] = wholesale_df['unit_price'] * wholesale_df['qty']
wholesale_df[['unit_price', 'qty', 'subtotal_line']].head()
```

```python
# Create a column for shipping price. Assume a shipping price of $7 per pound for orders over 50 pounds and $10 per pound for items 50 pounds or under.
def calculate_shipping_price(each_total_weight):
    if each_total_weight > 50:
        return each_total_weight * 7
    else: 
        return each_total_weight * 10

wholesale_df['total_weight'] = wholesale_df['unit_weight'] * wholesale_df['qty']
total_weight = wholesale_df["total_weight"]
total_weight.head()
tw = total_weight.tolist()
tw = [calculate_shipping_price(x) for x in tw]
#wholesale_df["shipping_price"] = wholesale_df[calculate_shipping_price (total_weight.all())]
#shipping_price = calculate_shipping_price(total_weight)
wholesale_df.insert(0, "shipping_price", tw, True)
wholesale_df.head()
```

```python
wholesale_df['total_weight'] 
# wholesale_df[['unit_price', 'unit_weight', 'qty', 'total_weight', 'shipping_price']].head()
```

```python
# Create a column for the total price
# using the subtotal and the shipping price
# along with a sales tax of 9.25%
wholesale_df['wholesale_subtotal'] = wholesale_df['unit_price'] * wholesale_df['qty']
wholesale_df['wholesale_price'] = ((wholesale_df['wholesale_subtotal'] + wholesale_df['shipping_price']) * 1.0925)
wholesale_df[['wholesale_subtotal', 'shipping_price', 'wholesale_price']].head().round(2)
```

```python
# Create a column for the cost
# of each line using unit cost, qty, and
# shipping price (assume the shipping cost
# is exactly what is charged to the client).
wholesale_df['wholesale_cost'] = (wholesale_df['unit_cost'] * wholesale_df['qty']) + wholesale_df['shipping_price']
wholesale_df.head()
```

```python
# Create a column for the profit of
# each line using line cost and line price
wholesale_df['wholesale_profit'] = (wholesale_df['wholesale_price'] - wholesale_df['wholesale_cost'])
wholesale_df.head()
```

## Part 4: Summarize and Analyze
Use the new columns with confirmed values to find the following information.

```python
# How much did each of the top 5 clients by quantity
# spend? Check your work from Part 1 for client ids.
top_clients = wholesale_df['client_id'].value_counts().head(5).index
top_clients = wholesale_df['wholesale_df['client_id].isin(top_clients)']
total_spending = top_clients.groupby('client_id')['line_price'].sum()
total_spending.index.name = none
total_spending.apply(lambda x: f'${x:.2f}')


```

```python
# Create a summary DataFrame showing the totals for the
# for the top 5 clients with the following information:
# total units purchased, total shipping price,
# total revenue, and total profit. Sort by total profit.
summary_df = wholesale_df[wholesale_df['client_id'].isin(top_clients)].groupby("client_id'").agg(
    qty_total = ('qty', 'sum'),
    shipping_total = ('shipping_price', 'sum'),
    revenue_total = ('line_price', 'sum'),
    cost_total = ('line_cost', 'sum')
    total_profit= ('line_profit', 'sum')
)
summary_df
```

```python
# Format the data and rename the columns
# to names suitable for presentation.
# Currency should be in millions of dollars.
display_df = summary_df.rename(columns={'client_id': 'Client ID',
                                        'qty_total': 'Units',
                                        'shipping_total': 'Shipping',
                                        'revenue_total': 'Total Revenue',
                                        'cost_total': 'Total Cost',
                                        'profit_total': 'Total Profit'})

converted_columns = ['Units', 'Shipping', 'Total Revenue', 'Total Cost', 'Total Profit']
for column in converted_columns:
    display_df[column] = ('$' + (display_df[column]).2f.astype(str))
print (display_df.to_string(index=True))
```

```python
# Sort the updated data by "Total Profit" form highest to lowest
display_df.sort_values("Total Profit", ascending = False)
```
